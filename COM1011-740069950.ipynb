    12. B: 1000(Bk - 0.63)^2 where Bk is the proportion of Black people by town

Target variable:
* MEDV: Median value of owner-occupied homes in $1000's

```python
# Answers here
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the datasets
train_data = pd.read_csv('mnist_train.csv')
test_data = pd.read_csv('mnist_test.csv')

# Separate features and labels
X_train = train_data.iloc[:, 1:].values  # All columns except the first one
y_train = train_data.iloc[:, 0].values    # First column (labels)

X_test = test_data.iloc[:, 1:].values    # All columns except the first one
y_test = test_data.iloc[:, 0].values      # First column (labels)

# Filter the data for only digits '0' and '1'
train_mask = (y_train == 0) | (y_train == 1)
X_train_binary = X_train[train_mask]
y_train_binary = y_train[train_mask]

test_mask = (y_test == 0) | (y_test == 1)
X_test_binary = X_test[test_mask]
y_test_binary = y_test[test_mask]

# Normalize the feature values
X_train_binary = X_train_binary / 255.0
X_test_binary = X_test_binary / 255.0

# Train the Perceptron
perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron.fit(X_train_binary, y_train_binary)

# Evaluate the Perceptron
test_accuracy = perceptron.score(X_test_binary, y_test_binary)
print(f'Test Accuracy: {test_accuracy:.6f}')

# Get predictions
y_pred_binary = perceptron.predict(X_test_binary)

# Create confusion matrix
conf_matrix = confusion_matrix(y_test_binary, y_pred_binary)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Calculate precision, recall, and F1 score
precision = precision_score(y_test_binary, y_pred_binary, pos_label=1)
recall = recall_score(y_test_binary, y_pred_binary, pos_label=1)
f1 = f1_score(y_test_binary, y_pred_binary, pos_label=1)

# Print results
print(f'Precision: {precision:.6f}')
print(f'Recall: {recall:.6f}')
print(f'F1 Score: {f1:.6f}')
```
